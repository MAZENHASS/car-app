## Outlier Treatment 

```python
for i in df.select_dtypes(include='number'):
    sns.boxplot(data= df, x= df[i])
    plt.show()
```

```python
df.head()
```

```python
def outliers(col): 
    q1 = col.quantile(0.25)
    q3 = col.quantile(0.75)
    iqr = q3 - q1
    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1 - 1.5 * iqr
    return upper_limit, lower_limit

# Copy df to avoid accidental data loss
df_clean = df.copy()

for i in ['Price', 'Levy', 'Prod. year', 'Engine volume', 'Mileage', 'Cylinders']:
    high, low = outliers(df_clean[i])
    
    outlier_mask = (df_clean[i] < low) | (df_clean[i] > high)
    outlier_count = outlier_mask.sum()
    total = df_clean.shape[0]
    percentage = round(100 * outlier_count / total, 2)
    
    print(f'Total outliers in "{i}": {outlier_count} ({percentage}%)')
    
    # Remove outliers
    df_clean = df_clean.loc[(df_clean[i] >= low) & (df_clean[i] <= high)]

# Optional: reset index
df_clean.reset_index(drop=True, inplace=True)
```

## Convert categorical LabelEncoder technique

```python

from sklearn.preprocessing import LabelEncoder

objects0 = df_clean.select_dtypes(include='object').copy()
nums0 = df_clean.select_dtypes(exclude='object').copy()

le = LabelEncoder()
for col in objects0.columns:
    objects0[col] = le.fit_transform(objects0[col])
    cg[col+'encoded']=objects0[col]
    

df_clean = pd.concat([objects0, nums0], axis=1)
```

## Model 

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import  PolynomialFeatures
from sklearn.svm import SVR 
from sklearn.tree import DecisionTreeRegressor 
from sklearn.ensemble import RandomForestRegressor 
from sklearn.ensemble import GradientBoostingRegressor 
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor, plot_tree, plot_importance

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score , mean_squared_error
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)

models = [
    LinearRegression(),
    Ridge(alpha=3),
    Lasso(alpha=2),
    DecisionTreeRegressor(),
    
    XGBRegressor(gamma=0,learning_rate=.05,n_estimators=200,subsample=.9,max_depth=7),
    RandomForestRegressor(n_estimators=20),
    GradientBoostingRegressor(),
    SVR(),
    KNeighborsRegressor()
]


algorithm =['LinearRegression','ridge','lasso','DecisionTreeRegressor','XGBRegressor','RandomForestRegressor','GradientBoostingRegressor','SVR','KNeighborsRegressor']
R2=[]
RMSE=[]
pre=[]
```

```python
x_train
```

```python
def evaluate_model(model, x_train=x_train, x_test=x_test, y_train=y_train, y_test=y_test):
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    pre.append(predictions)
    
    r2 = r2_score(y_test, predictions)
    rmse = np.sqrt(mean_squared_error(y_test, predictions))
    
    R2.append(r2)
    RMSE.append(rmse)
    
    print(f"Model: {model.__class__.__name__}")
    print(f"R² Score: {r2:.4f}")
    print(f"RMSE: {rmse:.4f}\n")
```

```python

for m in models:
    evaluate_model(m)
```

```python
dw=pd.DataFrame({'Algorithm': algorithm, 'R2score': R2,"RMSE":RMSE})
```
